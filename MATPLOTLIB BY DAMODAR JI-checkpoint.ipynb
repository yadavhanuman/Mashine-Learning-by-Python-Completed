```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4])
y= np.array([40,42,45,45])
plt.figure(figsize=(4,3))
plt.plot(x,y,marker='o',linestyle='dashed')
plt.grid()
plt.title('Report Card')
plt.xlabel('module')
plt.ylabel('marks')
plt.show()

```

```python
x= np.array([0,1,2,3,])
y=np.array([3,4,5,6])
plt.subplot(1,2,1)
plt.plot(x,y)
x= np.array([10,20,30,40])
y=np.array([3,4,5,6])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
module= np.array([1,2,3,4])
marks=np.array([40,42,45,45])
plt.scatter(module,marks)
plt.grid()
plt.show()
```

```python
a=np.array(['a','b','c','d'])
b=np.array([10,20,30,15])
plt.bar(a,b,color='c',width=0.1)
plt.figure(figsize=(4,3))
plt.show()
```

```python
g=np.array([35,40,45,70])
y=['BJP','CONG','AAP','SP']
plt.figure(figsize=(4,3))
plt.pie(g,labels=y,autopct='%0.2f%%')

plt.legend(
        
          title="Party",
          loc="center left",
          bbox_to_anchor=(1, 0.5))
plt.show()
```
