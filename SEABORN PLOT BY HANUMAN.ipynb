```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


```

# Line plot

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

x1= pd.DataFrame({'var':var,'var1':var1})
#figure=figsize=(4,1)
sns.lineplot(x='var',y='var1',data=x1)


 
```

```python
plt.plot(var,var1)

plt.show()
```

```python
y1=sns.load_dataset('penguins').head(20)
y1
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',
data=y1,hue='sex',style='sex',palette='Accent',markers=['o','*'],
dashes=False,legend='full')
plt.grid()
plt.title('python seaborn chart')
plt.show()
```

# Histogram Plot in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('penguins')
df
```

```python
sns.displot(df['flipper_length_mm'],
            bins=[170,180,190,200,218,220,238,240])
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],kde=True,rug=True,color='g',
            )
            
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df,)
```

```python
order=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',
order=order,hue_order=['Female','Male'],ci=100,n_boot=30)
```
